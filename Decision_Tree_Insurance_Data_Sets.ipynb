# Visualize the data sets.

In this data sets, looks like classification problem. yes/no problem
**Histograms**: Show the distribution of individual columns.

**Scatter Plots**: Show the relationship between each column and the bought_insurance variable.

**Box Plots**: Highlight the spread and outliers for each feature.

**Bar Plots**: Show the average bought_insurance for categorical features.

**Violin Plots**: Provide a combination of box plot and density plot for better visualization of data distribution across different categories.

**Pair Plot**: Provides a matrix of scatter plots to visualize pairwise relationships between features.

**Correlation Heatmap**: Shows the correlation coefficients between different features in a matrix form.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go

# Matplotlib and Seaborn plots
for plot in df.columns:
    if plot != 'bought_insurance':
        # Matplotlib - Histogram
        plt.figure(figsize=(10, 6))
        df[plot].hist(bins=80)
        plt.xlabel(plot)
        plt.ylabel('Frequency')
        plt.title(f'Histogram of {plot}')
        plt.show()

         # Plotly Histogram
        fig = px.histogram(df, x=plot, title=f'Histogram of {plot}')
        fig.show()

        # Seaborn - Histogram
        plt.figure(figsize=(10, 6))
        sns.histplot(df[plot], bins=80, kde=True)
        plt.xlabel(plot)
        plt.ylabel('Frequency')
        plt.title(f'Histogram of {plot}')
        plt.show()

        # Matplotlib - Scatter Plot
        plt.figure(figsize=(10, 6))
        plt.scatter(df[plot], df['bought_insurance'])
        plt.xlabel(plot)
        plt.ylabel('bought_insurance')
        plt.title(f'Scatter Plot of {plot} vs bought_insurance')
        plt.show()

         # Plotly Scatter Plot
        fig = px.scatter(df, x=plot, y='bought_insurance', title=f'Scatter Plot of {plot} vs bought_insurance')
        fig.show()

        # Seaborn - Scatter Plot
        plt.figure(figsize=(10, 6))
        sns.scatterplot(x=df[plot], y=df['bought_insurance'])
        plt.xlabel(plot)
        plt.ylabel('bought_insurance')
        plt.title(f'Scatter Plot of {plot} vs bought_insurance')
        plt.show()

        # Seaborn - Box Plot
        plt.figure(figsize=(10, 6))
        sns.boxplot(x=df[plot])
        plt.xlabel(plot)
        plt.title(f'Box Plot of {plot}')
        plt.show()

        # Plotly Box Plot
        fig = px.box(df, x=plot, title=f'Box Plot of {plot}')
        fig.show()

        # Seaborn - Bar Plot
        plt.figure(figsize=(10, 6))
        if df[plot].nunique() < 20:  # Ensuring the column has a reasonable number of unique values
            sns.barplot(x=df[plot], y=df['bought_insurance'], ci=None)
            plt.xlabel(plot)
            plt.ylabel('bought_insurance')
            plt.title(f'Bar Plot of {plot} vs bought_insurance')
            plt.show()
        else:
            print(f'Skipping bar plot for {plot} as it has too many unique values.')

        # Plotly Bar Plot
        if df[plot].nunique() < 20:  # Ensuring the column has a reasonable number of unique values
            fig = px.bar(df, x=plot, y='bought_insurance', title=f'Bar Plot of {plot} vs bought_insurance')
            fig.show()
        else:
            print(f'Skipping bar plot for {plot} as it has too many unique values.')

        # Seaborn - Violin Plot
        plt.figure(figsize=(10, 6))
        sns.violinplot(x='bought_insurance', y=df[plot], data=df)
        plt.xlabel('bought_insurance')
        plt.ylabel(plot)
        plt.title(f'Violin Plot of {plot} by bought_insurance')
        plt.show()

        # Plotly Violin Plot
        fig = px.violin(df, y=plot, x='bought_insurance', title=f'Violin Plot of {plot} by bought_insurance')
        fig.show()

        # Seaborn - Violin Plot
        plt.figure(figsize=(10, 6))
        sns.violinplot(x='bought_insurance', y=df[plot], data=df)
        plt.xlabel('bought_insurance')
        plt.ylabel(plot)
        plt.title(f'Violin Plot of {plot} by bought_insurance')
        plt.show()

        # Seaborn - Line Plot
        plt.figure(figsize=(10, 6))
        sns.lineplot(x=df.index, y=df[plot])
        plt.xlabel('Index')
        plt.ylabel(plot)
        plt.title(f'Line Plot of {plot}')
        plt.show()

        # Seaborn - KDE Plot
        plt.figure(figsize=(10, 6))
        sns.kdeplot(df[plot], shade=True)
        plt.xlabel(plot)
        plt.title(f'KDE Plot of {plot}')
        plt.show()

        # Seaborn - Joint Plot
        sns.jointplot(x=df[plot], y=df['bought_insurance'], kind="scatter")
        plt.suptitle(f'Joint Plot of {plot} vs bought_insurance', y=1.02)
        plt.show()

        # Seaborn - Regression Plot
        plt.figure(figsize=(10, 6))
        sns.regplot(x=df[plot], y=df['bought_insurance'])
        plt.xlabel(plot)
        plt.ylabel('bought_insurance')
        plt.title(f'Regression Plot of {plot} vs bought_insurance')
        plt.show()

# Seaborn - Pair Plot
sns.pairplot(df)
plt.suptitle('Pair Plot of DataFrame', y=1.02)
plt.show()

# Seaborn - Correlation Heatmap
plt.figure(figsize=(12, 10))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()


# Seaborn - Heatmap of entire DataFrame
plt.figure(figsize=(12, 10))
sns.heatmap(df, annot=True, cmap='coolwarm')
plt.title('Heatmap of DataFrame')
plt.show()

# Plotly Pair Plot
fig = px.scatter_matrix(df, title='Pair Plot of DataFrame')
fig.show()

# Plotly Correlation Heatmap
correlation_matrix = df.corr().reset_index().melt(id_vars='index')
fig = px.imshow(df.corr(), text_auto=True, title='Correlation Heatmap')
fig.show()



```

# Identify feature and target

```python
X = df[['age']]
y = df['bought_insurance']
```

# Preprocess the feature

```python
from sklearn.preprocessing import StandardScaler
x_scaled = StandardScaler().fit_transform(X)
x_scaled
```

# Divide the data for training and testing

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x_scaled,y, train_size = 0.80, random_state = 42)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
print(X_train.ndim, X_test.ndim, y_train.ndim, y_test.ndim)
```

# Create Model by training data sets and predict the unseen test data

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(criterion='entropy')
dtc.fit(X_train, y_train)
y_pred = dtc.predict(X_test)
```

# Check the model quality

```python
from sklearn.metrics import classification_report, confusion_matrix
report = classification_report(y_test, y_pred)
confusionmatrix = confusion_matrix(y_test, y_pred)
print(report)
print(confusionmatrix)
```
