```python
# BREAST CANCER WISCONSIN (DIAGNOSTIC) DATA SET
# A Simple Prediction
# Accuracy score : 0.9181286549707602

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# Reading the data from the file

pd.set_option('display.max_columns',None)
df=pd.read_csv('/content/Breast Cancer - data.csv',index_col=0)
```

```python
df.drop(['Unnamed: 32'], axis = 1 , inplace=True)
df.columns
```

```python
# The features that are needed for the training model
cols=['diagnosis',
      'radius_mean',
      'texture_mean',
      'perimeter_mean',
      'area_mean',
      'smoothness_mean',
      'compactness_mean',
      'concavity_mean',
      'concave points_mean',
      'symmetry_mean',
      'fractal_dimension_mean']
```

```python
x=df[cols]
x=x.drop('diagnosis',axis=1)
y=df['diagnosis']
```

```python
x.head().style.background_gradient(cmap = "twilight")
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
x_train.head().style.background_gradient(cmap = "Reds")
```

```python
x_test.head().style.background_gradient(cmap = "Blues")
```

```python
x_test.columns
```

```python
y_train
```

```python
y_test
```

```python
LRM =LogisticRegression()
LRM.fit(x_train,y_train)
```

```python
predictions=LRM.predict(x_test)
predictions
```

```python
x_test.index[:]
```

```python
output = pd.DataFrame({'Id': x_test.index[:], 'Diagnosis': predictions})
output.head(20).style.background_gradient(cmap = "Reds")
```

```python
output.to_csv('predictions.csv', index=False)
print("Predictions was successfully saved!")
```

```python
accuracy_score(y_test,predictions)
```
